## Чтение данных

```python
# Put your code here
data = dep36['item 95519']
```

```python
plt.figure(figsize=(20, 20))
plt.plot(data)
plt.show()
```

Данные выглядят очень стохастическими, похожими на белый шум. Сезонности нет.

```python
from sklearn.model_selection import train_test_split
```

```python
train_data = data[0:int(data.size * 0.8)]
test_data = data[train_data.size:]
```

За исключением некоторых выбросов, дисперсия временного ряда не меняется в зависимости от времени. Поэтому преобразование Бокса-Кокса не требуется.

Проверим ряд на нестационарность с помощью критериев КПСС и Дики-Фуллера.

```python
from statsmodels.tsa import stattools

kpss = stattools.kpss(train_data)

print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss[0], kpss[1], kpss[2], kpss[3]))
```

```python
adfuller_stat = stattools.adfuller(train_data)

print("Adfuller pvalue {}".format(adfuller_stat[1]))
```

Гипотеза о стационарности уверенно отвергается на уровне значимости 0.01. 
Гипотеза критерия Дики-Фуллера не отвергается, что также говорит о нестационарности.

Попробуем привести ряд к стационарному путем удаления выбросов, так как выбросы останутся и в продифференцированном ряду.

```python
cleared_data = train_data.loc["2005-06-15": ]

kpss_stat = stattools.kpss(cleared_data)
adfuller_stat = stattools.adfuller(cleared_data)

print("KPSS test pvalue {}".format(kpss_stat[1]))
print("Adfuller test pvalue {}".format(adfuller_stat[1]))
```

Теперь гипотеза о стационарности не отвергается, а гипотеза критерия Дики-Фуллера отвергается, что позволяет говорить о стационарности ряда.

Найдем начальные приближения для поиска оптимальной модели ARIMA с параметром d=0, так как дифференцирование не потребовалось.

```python
import statsmodels.api as sms
```

```python
sms.graphics.tsa.plot_acf(cleared_data.values.squeeze(), lags=15)
```

```python
sms.graphics.tsa.plot_pacf(cleared_data.values.squeeze(), lags=15)
```

Исходя из графиков ACF и PACF, выберем начальные приближения:

для параметра p=1 (номер последнего значимого лага частичной автокорреляции, за исключением слишком больших. 

для параметра q=1 (номер последнего значимого лага автокорреляции)

В модели предполагается отсутствие сезонности, поэтому параметры P и Q равны нулю.

Порядок дифференцирования d = 0

```python
ps = range(0, 14)
qs = range(0, 14)
d = 0
```

```python
from itertools import product

parameters = product(ps, qs)
parameters_list = list(parameters)
print(len(parameters_list))
```

```python
results = []
best_aic = float("inf")

for param in parameters_list:
    try:
        model=sms.tsa.statespace.SARIMAX(train_data.loc["2005-06-15": ], order=(param[0], d, param[1])).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
best_param
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

Лучшая модель - ARIMA(4, 0, 1)

## Анализ остатков

```python
plt.figure(figsize=(15,8))
best_model.resid.plot()
plt.ylabel(u'Residuals')
```

```python
import scipy.stats as sts

plt.figure(figsize=(15,9))
plt.subplot(211)
best_model.resid.plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sms.graphics.tsa.plot_acf(best_model.resid.values.squeeze(), lags=50, ax=ax)

print("Критерий Стьюдента: p=%f" % sts.ttest_1samp(best_model.resid, 0)[1])
print("Критерий КПСС: p=%f" % sms.tsa.stattools.kpss(best_model.resid)[1])
```

Остатки несмещены, неавтокоррелированы и имеют одинаковую дисперсию. Исходя из этого можно предположить, что модель неплохо подходит к данным.

* 1) plot the graph of the train TS, preprocess data (delete tail, make calendar transformation if needed)
* 2) stabilize dispersion of train TS through box-cox transformation if needed.
* 3) Run grid search for ARIMA parameters (use transformed train TS) according to AIC. Use at least 10 distinct values for $p,q,d$. How many time does it take?
* 4) Build forecast for the initial TS with best ARIMA model selected automaticaly at step 2 (you need to transform it like in step 2).

Для поиска параметров воспользуемся данными из предыдущего пункта.

```python
ps = range(0, 10)
qs = range(0, 10)
ds = range(0, 10)

parameters = product(ps, ds, qs)
parameters_list = list(parameters)
print(len(parameters_list))
```

```python
from tqdm import tqdm
```

```python
results = []
best_aic = float("inf")

for param in tqdm(parameters_list):
    try:
        model=sms.tsa.statespace.SARIMAX(train_data.loc["2005-06-15": ], order=(param[0], param[1], param[2])).fit(
            disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
best_param
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
forecast = best_model.predict(start=284, end=(284+80-1))

loss = qualityMACAPE(forecast, test_data.values)[0]
print(loss)
```
